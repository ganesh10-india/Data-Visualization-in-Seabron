# Plotting with categorical dataÂ¶


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='ticks',color_codes= True)
```

# Categorical scatterplots.

```python
tips = sns.load_dataset("tips")
sns.catplot(x = 'day',y='total_bill',data = tips)
```

```python
# The jitter parameter controls the magnitude of jitter or disables it altogether:

sns.catplot(x= 'day',y='total_bill',jitter = False,data = tips)
```

```python
# The second approach adjusts the points along the categorical axis using an algorithm that prevents them from overlapping

sns.catplot(x='day',y='total_bill',kind ='swarm',data = tips)
```

```python
# itâs possible to add another dimension to a categorical plot by using a hue semantic.

sns.catplot(x= 'day',y='total_bill',hue= 'sex',kind ='swarm',data =tips)
```

```python
sns.catplot(x= 'size',y='total_bill',kind ='swarm',data = tips.query("size ! = 3"))
```

```python
sns.catplot(x='smoker',y ='tip',order=['No','Yes'],data =tips)
```

```python
sns.catplot(x='total_bill',y='day',hue='time',kind='swarm',data =tips)
```

# Distributions of observations within categories


# Boxplot

```python
sns.catplot(x='day',y='total_bill',kind = 'box',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',hue='smoker',kind='box',data =tips)
```

```python
tips['weekend'] = tips['day'].isin(['Sat','Sun'])
sns.catplot(x='day',y='total_bill',hue='weekend',kind ='box',dodge= False,data = tips)
```

```python
# boxenplot(), draws a plot that is similar to a box plot but optimized for showing more information about the shape of the distribution. It is best suited for larger datasets

diamonds = sns.load_dataset('diamonds')
sns.catplot(x='color',y='price',kind='boxen',data =diamonds.sort_values("color"))
```

# Violinplots

```python
sns.catplot(x='total_bill',y='day',hue='sex',kind='violin',data = tips)
```

```python
sns.catplot(x='total_bill',y='day',hue= 'sex',kind= 'violin',bw=.15,cut= 0,data=tips)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',split= True,data =tips)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',inner='stick',split=True,palette='pastel',data= tips)
```

```python
g = sns.catplot(x='day',y='total_bill',kind='violin',inner =None,data =tips)
sns.swarmplot(x='day',y='total_bill',color= 'k',size=3,data=tips,ax= g.ax)
```

# Statistical estimation within categories


# Bar plot

```python
titanic = sns.load_dataset('titanic')
sns.catplot(x='sex',y='survived',hue='class',kind= 'bar',data =titanic)
```

```python
sns.catplot(x='deck',kind='count',palette="ch:.25",data=titanic)
```

```python
sns.catplot(y='deck',hue='class',kind='count',palette='pastel',edgecolor='0.6',data =titanic)
```

# Point plots

```python
sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic);
```

```python
sns.catplot(x="class", y="survived", hue="sex",
            palette={"male": "g", "female": "m"},
            markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=titanic);
```

# Showing multiple relationships with facets

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.6,
            kind="swarm", data=tips);
```

```python
g = sns.catplot(x="fare", y="survived", row="class",
                kind="box", orient="h", height=1.5, aspect=4,
                data=titanic.query("fare > 0"))
g.set(xscale="log");
```
